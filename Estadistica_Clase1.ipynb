```
# edad minima
#titanic['age'].min()

m = np.min(titanic['age'])
M = np.max(titanic.age)

bins = []

for i in range(int(m), int(M)+1, 10):
  bins.append(i)

#print(bins)

titanic['agebin'] = pd.cut(titanic.age, bins)

titanic
```

```
titanic[titanic.survived ==1]['agebin'].value_counts().sort_index().plot(kind = 'bar')
plt.show()
```

```
# diagramas de tipo kde : kernel density estimation
titanic.age[titanic.pclass == 3]

fig = plt.figure(figsize= (20,10))
for clases in [1,2,3]:
  titanic.age[titanic.pclass == clases].plot(kind = 'kde')

plt.legend({"3st Class", "2nd Class" , "1th Class"})
plt.show()
```

```
titanic.age[titanic.pclass == 1].plot(kind = 'kde')
plt.title("KDE : Edad vs. Clase")
plt.show()
```
