```
# Tercer Ejercicio dinamico

#Erick Bedoya


vector = np.array([18.1, 18.5, 18.7, 19.9, 20.4, 20.5, 21.1, 21.3, 21.3, 21.3,21.3 ,22.0, 22.2, 22.4])
asimetria = 3 * ((np.mean(vector) - np.median(vector))/ np.std(vector))
print(asimetria)
```

```
# Diana Rodriguez

cant_ventas_leche = pd.Series([18.1, 18.5, 18.7, 19.9, 20.4,20.5, 21.1, 21.3, 21.3, 21.3, 21.3,22,22.2,22.4], index = ["Enero 1", "Febrero 1", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto", "Setiembre", "Octubre", "Noviembre","Diciembre","Enero 2", "Febrero 2" ])
coeficienteAsimetria = ((cant_ventas_leche.mean()- cant_ventas_leche.median())*3)/np.std(cant_ventas_leche)
coeficienteAsimetria
```
